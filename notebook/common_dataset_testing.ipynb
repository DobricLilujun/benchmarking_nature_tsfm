```python

path = "/home/snt/projects_lujun/benchmarking_nature_tsfm/data/M/M1/arrexp.dat"
import math
import re
from collections import defaultdict
import pandas as pd

sci_pattern = re.compile(r'[+-]?\d+\.\d+E[+-]?\d+')

records = defaultdict(dict)

with open(path, "r") as f:
    current_key = None
    current_meta = None
    current_values = []
    
    for line in f:
        line = line.strip()
        if not line:
            continue
        
        # 判断是不是头行（YAF开头）
        if line.startswith("YAF"):
            # 如果前一个 key 已经有数据，保存
            if current_key:
                records[current_key] = {
                    "meta": current_meta,
                    "values": current_values
                }
            
            parts = line.split()
            current_key = parts[0]      # 记录名
            current_meta = parts[1:]    # 后面几个数字
            current_values = []         # 清空数值缓存
        
        else:
            # 数据行，正则提取所有科学计数数字
            nums = sci_pattern.findall(line)
            current_values.extend([float(x) for x in nums])
    
    # 别忘了最后一条
    if current_key:
        records[current_key] = {
            "meta": current_meta,
            "values": current_values
        }

# 转成 DataFrame（可选）
df = pd.DataFrame([
    {"key": k, "meta": v["meta"], "values": v["values"]}
    for k, v in records.items()
])

print(df.head())
```
