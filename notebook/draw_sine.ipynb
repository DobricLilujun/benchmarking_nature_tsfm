```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2 * np.pi, 2 * np.pi, 500)

y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('正弦函数 sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)  # 两个周期
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 'r-', linewidth=10)  # 红色加粗曲线

# 移除默认坐标轴
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['left'].set_position('zero')

# 横纵轴加粗
plt.gca().spines['bottom'].set_linewidth(10)
plt.gca().spines['left'].set_linewidth(10)

# 在坐标轴末尾添加箭头
plt.arrow(2 * np.pi, 0, 0.5, 0, head_width=0.9, head_length=1.0, fc='black', ec='black')  # 横轴箭头
plt.arrow(0, 1.2, 0, 0.5, head_width=0.9, head_length=1.0, fc='black', ec='black')  # 纵轴箭头

# 关闭刻度线和标签
plt.xticks([])
plt.yticks([])
plt.savefig('plot_transparent.png', transparent=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_sine_with_dashed_fraction(fraction_of_period=0.5, total_periods=2):
    """
    Plot sine wave for given total_periods. The initial part up to fraction_of_period (in one period) is a solid line;
    the rest is dashed line.
    
    Parameters:
    - fraction_of_period: fraction of one period shown as solid line (e.g., 0.5 means half period solid)
    - total_periods: total number of sine wave periods to plot
    """
    x = np.linspace(0, total_periods * 2 * np.pi, 500)
    y = np.sin(x)

    period_len = 2 * np.pi
    split_x = fraction_of_period * period_len  # x position where solid ends and dashed begins in one period

    plt.figure(figsize=(8, 4))

    # Plot solid line up to fraction_of_period * one period (but over total periods, all periods scaled)
    # So calculate real split point in x domain:
    solid_end = split_x
    # Because we have total_periods, solid applies only for the first fraction_of_period of the first period
    # The rest is plotted as dashed
    
    # Plot solid part up to solid_end
    mask_solid = x <= solid_end
    plt.plot(x[mask_solid], y[mask_solid], 'r-', linewidth=10)

    # Plot dashed part from solid_end to end of total_periods * 2pi
    mask_dashed = x > solid_end
    plt.plot(x[mask_dashed], y[mask_dashed], 'r--', linewidth=10)

    # Remove default top and right spines
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    # Position bottom and left spines at zero
    plt.gca().spines['bottom'].set_position('zero')
    plt.gca().spines['left'].set_position('zero')

    # Make horizontal and vertical axis thicker
    plt.gca().spines['bottom'].set_linewidth(10)
    plt.gca().spines['left'].set_linewidth(10)
    plt.savefig(f"plot_transparent_{fraction_of_period}_{total_periods}.png", transparent=True)

    # Add arrows on axes ends
```
